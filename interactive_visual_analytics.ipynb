```python
marker_cluster = MarkerCluster()
```

```python
def assign_marker_color(row):
    if row['class'] ==1:
        return 'green'
    elif row['class'] ==0:
        return 'red'
    else:
        return 'uknown'

df_spcx['marker_color'] = df_spcx.apply(assign_marker_color, axis=1)
```

```python
new_map.add_child(marker_cluster)
for index, row in df_spcx.iterrows():
    coordinate = [row['Lat'], row['Long']]
    folium.map.Marker(coordinate, icon=folium.Icon(color='white', icon_color= row['marker_color'])).add_to(marker_cluster)
new_map
```

```python
from math import sin, cos, sqrt, atan2, radians
```

```python
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1=radians(lat1)
    lon1=radians(lon1)
    lat2=radians(lat2)
    lon2=radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a),sqrt(1-a))

    distance = R * c
    return distance
```

```python
coastline = [28.5656,-80.56847 ]
lat1 = slc_40[0][0]
lon1 = slc_40[0][1]
lat2 = coastline[0]
lon2 = coastline[1]

distance = calculate_distance(lat1,lon1,lat2,lon2)
slc_40_cor = [lat1,lon1]
```

```python

distance_marker = folium.Marker(
    coastline,
    icon = DivIcon(
        icon_anchor=(0,0),
        icon_size=(20,20),
        html= '<div style = "font-size: 12; color:#d35400;"><b>%s</b><.div>' % "{:10.2f} KM".format(distance),
    )
)
new_map.add_child(distance_marker)
```
